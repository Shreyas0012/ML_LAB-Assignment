Analyze the Titanic dataset to understand factors affecting passenger survival using data preprocessing and exploratory data analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()

```

```python
df['Cabin'] = np.nan
df['Cabin'] = df['Cabin'].astype(object)
df.loc[df.index % 5 != 0, 'Cabin'] = 'C' + df.index[df.index % 5 != 0].astype(str)

df.loc[df.sample(frac=0.1).index, 'Age'] = np.nan

import random
ports = ['S', 'C', 'Q', np.nan]
df['Embarked'] = [random.choice(ports) for _ in range(len(df))]
df.loc[df.sample(frac=0.05).index, 'Embarked'] = np.nan

df.to_csv('titanic_with_missing.csv', index=False)
print("Generated 'titanic_with_missing.csv' with introduced missing values.")
print("You can now load this new file to see the missing values.")
```

```python
df = pd.read_csv('titanic_with_missing.csv')
df.head()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.show()
```
